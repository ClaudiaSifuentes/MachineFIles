```python
def calculate_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
    outlier_percentage = (outliers / df.shape[0]) * 100
    return outliers, outlier_percentage

outlier_counts, outlier_percentages = calculate_outliers(clean_df.select_dtypes(include=[np.number]))

outlier_summary = pd.DataFrame({
    'Outlier Count': outlier_counts,
    'Outlier Percentage': outlier_percentages
})

print(outlier_summary)
```

```python
# Step 1: Define a function to remove outliers
def remove_outliers(df):
    numeric_cols = df.select_dtypes(include=['number'])
    Q1 = numeric_cols.quantile(0.25)
    Q3 = numeric_cols.quantile(0.75)
    IQR = Q3 - Q1
    return df[~((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).any(axis=1)]

# Step 2: Iteratively remove outliers until no more are detected
clean_df_no_outliers = clean_df.copy()
previous_shape = clean_df_no_outliers.shape
while len(clean_df_no_outliers) > 500:
    clean_df_no_outliers = remove_outliers(clean_df_no_outliers)
    if clean_df_no_outliers.shape == previous_shape:
        break
    previous_shape = clean_df_no_outliers.shape

# Display the first 20 rows of the cleaned DataFrame
clean_df_no_outliers.head(20)
```
